```python
import matplotlib.pyplot as plt
import numpy as np

total = 15

human1 = 13
gpt1 = 9
claude1 = 11
gemini1 = 9

human2 = 11
gpt2 = 11
claude2 = 6
gemini2 = 8

human3 = 13
gpt3 = 5
claude3 = 8
gemini3 = 8

subsets = ("Subset1", "Subset2", "Subset3")
actual = [total, total, total]
human = [human1, human2, human3]
gpt = [gpt1, gpt2, gpt3]
claude = [claude1, claude2, claude3]
gemini = [gemini1, gemini2, gemini3]

bar_width = 0.15
x = np.arange(len(subsets))

plt.bar(x - 2*bar_width, actual, bar_width, label='Actual')
plt.bar(x - bar_width, human, bar_width, label='Human Control')
plt.bar(x, gpt, bar_width, label='Open AI - GPT 4 Vision')
plt.bar(x + bar_width, claude, bar_width, label='Anthropic - Claude 3 Opus')
plt.bar(x + 2*bar_width, gemini, bar_width, label='Google - Gemini Pro 1.0 Vision')

plt.xlabel('Subsets')
plt.ylabel('Accuracy')
plt.title('Comparison of Different Models')

plt.xticks(x, subsets)
# plt.legend()
plt.tight_layout()
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), fancybox=True, shadow=True, ncol=2)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (10,6)
plt.figure(facecolor='white')

total = 15

human1 = 13
gpt1 = 9
claude1 = 11
gemini1 = 9

human2 = 11
gpt2 = 11
claude2 = 6
gemini2 = 8

human3 = 13
gpt3 = 5
claude3 = 8
gemini3 = 8

subsets = ("Subset1", "Subset2", "Subset3")
actual = [total, total, total]
human = [human1, human2, human3]
gpt = [gpt1, gpt2, gpt3]
claude = [claude1, claude2, claude3]
gemini = [gemini1, gemini2, gemini3]

bar_width = 0.15
x = np.arange(len(subsets))

plt.bar(x - 2*bar_width-0.04, actual, bar_width, label='Total Characters')
plt.bar(x - bar_width-0.02, human, bar_width, label='Human Control')
plt.bar(x, gpt, bar_width, label='Open AI - GPT 4 Vision')
plt.bar(x + bar_width+0.02, claude, bar_width, label='Anthropic - Claude 3 Opus')
plt.bar(x + 2*bar_width+0.04, gemini, bar_width, label='Google - Gemini Pro 1.0 Vision')

plt.xlabel('Subset of Notes Used')
plt.ylabel('Accuracy Raw Score')
plt.suptitle('Comparison of Accuracy of Different Off-the-Shelf Image Recognition Models',fontsize=16)
plt.xticks(x, subsets)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), fancybox=True, shadow=True, ncol=2, prop={'size': 14})
plt.tight_layout()

# Add percentage labels on top of each bar
for i in range(len(x)):
    plt.text(x[i] - 2*bar_width-0.04, actual[i] + 0.2, f"{int((actual[i] / human[i]) * 100)}%", ha='center')
    plt.text(x[i] - bar_width-0.02, human[i] + 0.2, "100%", ha='center')
    plt.text(x[i], gpt[i] + 0.2, f"{int((gpt[i] / human[i]) * 100)}%", ha='center')
    plt.text(x[i] + bar_width+0.02, claude[i] + 0.2, f"{int((claude[i] / human[i]) * 100)}%", ha='center')
    plt.text(x[i] + 2*bar_width+0.04, gemini[i] + 0.2, f"{int((gemini[i] / human[i]) * 100)}%", ha='center')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator

plt.rcParams["figure.figsize"] = (8, 3)
plt.figure(facecolor='white')

total = 15

human1 = 13
gpt1 = 9
claude1 = 11
gemini1 = 9

human2 = 11
gpt2 = 11
claude2 = 6
gemini2 = 8

human3 = 13
gpt3 = 5
claude3 = 8
gemini3 = 8

subsets = ("Subset1", "Subset2", "Subset3")
actual = [total, total, total]
human = [human1, human2, human3]
gpt = [gpt1, gpt2, gpt3]
claude = [claude1, claude2, claude3]
gemini = [gemini1, gemini2, gemini3]

data = [human, gpt, claude, gemini]
labels = ['Human Control', 'GPT 4 Vision', 'Claude 3 Opus', 'Gemini Pro 1.0 Vision']

plt.boxplot(data, labels=labels)

plt.xlabel('Model')
plt.ylabel('Accuracy Raw Score')
plt.title('Box and Whisker Plot of Different Off-the-Shelf Image Recognition Models')

plt.gca().yaxis.set_major_locator(MultipleLocator(2))  # Set major ticks every 2 units

plt.show()
```
